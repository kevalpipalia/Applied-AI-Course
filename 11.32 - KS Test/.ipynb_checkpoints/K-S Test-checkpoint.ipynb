```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
x = stats.norm.rvs(size = 1000)
sns.set_style('whitegrid')
sns.kdeplot(np.array(x),bw=0.5)
plt.show()
```

```python
stats.kstest(x,'norm')
```

### NOTE:
    Here the p-value is large i.e. > 5% so we accept the null hypothesis means we can say that both are same distribution.

```python
y = np.random.uniform(0,1,10000)
sns.kdeplot(np.array(y), bw=0.1)
plt.show()
```

```python
stats.kstest(y, 'norm')
```

### NOTE:
here as we can see the p-value is too low, so we can say alternative hypothesis is true i.e. both are different distributions.
